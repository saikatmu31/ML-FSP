```python
url = 'http://bit.ly/imdbratings'
movies = pd.read_csv(url)
movies.head()
```

```python
movies.title.sort_values(ascending=False,inplace=False)
movies.head()
```

```python
movies.head()
```

```python
movies.sort_values(by="duration",ascending=False,inplace=True)
movies.head()
```

```python
columns = ['content_rating','duration']
movies.sort_values(by = columns)
```

```python
bools = []

for length in movies.duration:
    if length>=200:
        bools.append(True)
    else:
        bools.append(False)
        
bools[0:10]
```

```python
len(bools)
```

```python
is_long = pd.Series(bools)
```

```python
is_long.head()
```

```python
movies[is_long]
```
